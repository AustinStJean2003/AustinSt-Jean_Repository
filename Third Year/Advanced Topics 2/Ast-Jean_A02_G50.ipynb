```python
import pandas as pd
df = pd.DataFrame(pd.read_csv('Death_rates_for_suicide__by_sex__race__Hispanic_origin__and_age__United_States.csv'))
df
```

Read data from CSV

```python
filtereddf = df[["STUB_NAME", "STUB_LABEL", "YEAR", "AGE", "ESTIMATE"]]
filtereddf
```

Filter out unneeded data

```python
filtereddf = filtereddf.dropna()
filtereddf
```

Remove nans from dataset

```python
filtereddf = filtereddf.copy()
filtereddf.rename(columns={"STUB_NAME": "CATEGORY"}, inplace=True)
filtereddf.rename(columns={"STUB_LABEL": "PEOPLE"}, inplace=True)
filtereddf
```

Renaming columns to be more representational of their data

```python
filtereddf[">15"] = filtereddf["ESTIMATE"] > 15
filtereddf
```

Add a column that displays true or false based on if estimate is greater than 15

```python
everyone = filtereddf[filtereddf["PEOPLE"] == "All persons"]
highestEstimate = everyone.ESTIMATE.max()
highestEstimateRow = everyone.loc[everyone["ESTIMATE"] == highestEstimate]
highestEstimateRow = highestEstimateRow[["YEAR", "ESTIMATE"]]
highestEstimateRow

```

Question 1: Find the year with the highest overall estimate when People column is all persons

```python
lowestEstimate = everyone.ESTIMATE.min()
lowestEstimateRow = everyone.loc[everyone["ESTIMATE"] == lowestEstimate]
lowestEstimateRow = lowestEstimateRow[["YEAR", "ESTIMATE"]]
lowestEstimateRow.drop_duplicates(inplace=True)
lowestEstimateRow
```

Question 2: Find the year with the lowest overall estimate by People column is all persons

```python
maledf = filtereddf[filtereddf["PEOPLE"].str.contains("Male")]
highestMaleEstimate = maledf.ESTIMATE.max()
highestMaleRow = maledf.loc[maledf["ESTIMATE"] == highestMaleEstimate]
highestMaleRow = highestMaleRow[["YEAR", "ESTIMATE"]]
highestMaleRow
```

Question 3: What year has the highest estimate for males

```python
femaledf = filtereddf[filtereddf["PEOPLE"].str.contains("Female")]
highestFemaleEstimate = femaledf.ESTIMATE.max()
highestFemaleRow = femaledf.loc[femaledf["ESTIMATE"] == highestFemaleEstimate]
highestFemaleRow = highestFemaleRow[["YEAR", "ESTIMATE"]]
highestFemaleRow
```

Question 4: What year has the highest estimate for females

```python
averageEstimate = round(filtereddf["ESTIMATE"].mean(), 2)
averageEstimate
```

Question 5: What is the average number overall across all years

```python
averageEstimatesByAge = filtereddf.groupby("AGE")["ESTIMATE"].mean()
highestAverageAgeGroup = averageEstimatesByAge.idxmax()
highestAverage = averageEstimatesByAge.max()
highestAverage = round(highestAverage, 2)
print("Age group with the highest Estimate is", highestAverageAgeGroup, "with an average of", highestAverage)
```

Question 6: Which age group has the highest average estimate?

```python
lowestAverageAgeGroup = averageEstimatesByAge.idxmin()
lowestAverage = averageEstimatesByAge.min()
lowestAverage = round(lowestAverage, 2)
print("Age group with the lowest Estimate is", lowestAverageAgeGroup, "with an average of", lowestAverage)
```

Question 7: Which age group has the lowest estimate?

```python
maleGt15 = filtereddf[filtereddf["PEOPLE"].str.contains("Male")][">15"].sum()
femaleGt15 = filtereddf[filtereddf["PEOPLE"].str.contains("Female")][">15"].sum()
print("Count of Estimates by Gender")
print("Male total:", maleGt15, "Female total:", femaleGt15)
```

Question 8: How many male and female estimates are greater than 15

```python
import matplotlib.pyplot as plt
sortedmaletop10 = maledf.sort_values(by="AGE", ascending=False)
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])
ax.barh(sortedmaletop10['AGE'], sortedmaletop10['ESTIMATE'], color="blue")
plt.ylabel('AGE')
plt.xlabel('ESTIMATE')
plt.title("Top Entries By Age for Males")
plt.show()
```
